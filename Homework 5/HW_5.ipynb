## (a) Estimate the mean annual evaporation (E, mm/year) for each watershed

The equations for the long-term water balance is 
$P = E + Q$

Where, $P$ = precipitation, $E$ = evaporation and $Q$ = runoff.

For estimating the mean annual evaporation (E, mm/year) for each watershed:

$E = P - Q$

```python
df['E'] = df['P'] - df['Q']
```

```python
df.head()
```
