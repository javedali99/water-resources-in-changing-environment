```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

## Qusetion 3 (a): Plot surface runoff (Q) versus rainfall (P)

```python
# create some values for p (rainfall)
p = np.array(range(0, 500))
```

```python
q1 = p
```

```python
def calculate_q(S, a, m):
    Q = (((a-1)*p) - S * (np.sqrt((m+1)**2 - 2*a*m)) + (np.sqrt((p + (m+1)*S)**2 - 2*a*m*S*S - 2*a*S*p)))/1.9
    return Q
```

```python
Q2 = calculate_q(50, 1.9, 0.2025)
```

```python
Q3 = calculate_q(50, 1.9, 0)
```

```python
Q4 = calculate_q(300, 1.9, 0.2025)
```

```python
Q5 = calculate_q(300, 1.9, 0)
```

```python
plt.style.use("seaborn")
plt.figure(figsize=(14, 10))

plt.plot(q1, label="$S_b = 0$")
plt.plot(Q2, label="$S_b = 50$, $\Psi = 0.2$, $a = 1.9$")
plt.plot(Q3, label="$S_b = 50$, $\Psi = 0$, $a = 1.9$")
plt.plot(Q4, label="$S_b = 300$, $\Psi = 0.2$, $a = 1.9$")
plt.plot(Q5, label="$S_b = 300$, $\Psi = 0$, $a = 1.9$")


plt.title("Surface Runoff ($Q$) versus Rainfall ($P$)", fontsize=18, y=1.03)

plt.xlabel("Rainfall", fontsize=14)
plt.ylabel("Surface runoff", fontsize=14)

plt.legend(loc='best', fontsize=12)

plt.savefig("hw4_1_a.png", dpi=300)

plt.show()
```
